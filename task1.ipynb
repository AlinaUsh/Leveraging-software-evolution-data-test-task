```python
def is_subsequence(old_version, new_version):
    i, j = 0, 0
    while i < len(old_version) and j < len(new_version):
        if old_version[i] == new_version[j]:
            i += 1
        j += 1
    return i == len(old_version)
```

```python
df[df['subject'] != df['message']][['subject', 'message']].sample(3)
```

```python
def normalize_str(s):
    s = ' '.join(s.split()).lower().strip()
    return s

print(f"message != subject in {len(df[df['subject'].apply(normalize_str) != df['message'].apply(normalize_str)]) / len(df) * 100}% samples")

subject_in_message = [is_subsequence(normalize_str(row['subject']), normalize_str(row['message'])) for _, row in df.iterrows()]
print(f"subject is a subsequence of message in {len(subject_in_message) / len(df) * 100}% samples")
subject_in_message = [normalize_str(row['subject']) in normalize_str(row['message']) for _, row in df.iterrows()]
print(f"subject is a substring of message in {len(subject_in_message) / len(df) * 100}% samples")
```

```python
added_only = [is_subsequence(row['old_contents'], row['new_contents']) for _, row in df.iterrows()]
deleted_only = [is_subsequence(row['new_contents'], row['old_contents']) for _, row in df.iterrows()]

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot for added_only
sns.countplot(x=added_only, ax=axes[0])
axes[0].set_title('Added Only')
axes[0].set_xlabel('Is Subsequence')
axes[0].set_ylabel('Count')

# Plot for deleted_only
sns.countplot(x=deleted_only, ax=axes[1])
axes[1].set_title('Deleted Only')
axes[1].set_xlabel('Is Subsequence')
axes[1].set_ylabel('Count')

plt.show()
```

## Models + Evaluation

```python
checkpoint = "smallcloudai/Refact-1_6B-fim"
device = "cuda" # for GPU usage or "cpu" for CPU usage

tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForCausalLM.from_pretrained(checkpoint, trust_remote_code=True).to(device)

prompt = '<fim_prefix>def print_hello_world():\n    """<fim_suffix>\n    print("Hello world!")<fim_middle>'

inputs = tokenizer.encode(prompt, return_tensors="pt").to(device)
outputs = model.generate(inputs, max_length=100, temperature=0.2)
print("-"*80)
print(tokenizer.decode(outputs[0]))
```
